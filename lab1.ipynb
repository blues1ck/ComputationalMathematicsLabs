Здесь определил сами функции, с которыми буду работать

```python
import math
import matplotlib.pyplot as plt

x_0 = 1.5

def function1(x):
    return math.sin(x*x)

def function2(x):
    return math.cos(math.sin(x))

def function3(x):
    return math.exp(math.sin(math.cos(x)))

def function4(x):
    return math.log(x+3)

def function5(x):
    return (x+3)**(0.5)
```

Здесь определил формулы для подсчета производных

```python
def f1(function, x, h):
    return (function(x + h) - function(x))/h

def f2(function, x, h):
    return (function(x) - function(x - h))/h

def f3(function, x, h):
    return (function(x + h) - function(x - h))/(2*h)

def f4(function, x, h):
    return 4*f3(function, x, h)/3 - (function(x + 2*h)-function(x - 2*h))/(3*4*h)

def f5(function, x, h):
    return 3*f3(function, x, h)/2 - 3*(function(x + 2*h)-function(x - 2*h))/(5*4*h) + (function(x + 3*h) - function(x - 3*h))/(10*6*h)

f = [f1, f2, f3, f4, f5]
```

Здесь создаю функцию для построения графика зависимости (f'_аналитическая - f'_приближенная)(h)

```python
def create_graph(axes, name):
    h = [math.log(axes[0][i]) for i in range(len(axes[0]))]
    plt.figure()
    for i in range(1, 6):
        plt.plot(h, axes[i], label = f"ln(|f' - f'{i}|)")
    plt.title(f"График зависимости модуля абсолютной погрешности от шага в логарифмическом масштабе для функции {name}")
    plt.xlabel("ln(h)")
    plt.ylabel("ln(|f'-f'i|)")
    plt.grid()
    plt.legend()
    
```

Функция sin(x^2). Аналитически её производная (sin(x^2))' = 2x*cos(x^2).

```python
def function1_derivative(x):
    return (2*x*math.cos(x*x))

axes = [[] for _ in range(6)] #матрица результатов для графика

for n in range(1, 21):
    h = 2 / (2**n)
    axes[0].append(h)
    resultInPoint = function1_derivative(x_0)
    for i in range(0, 5):
        axes[i + 1].append(math.log(abs(f[i](function1, x_0, h) - resultInPoint)))

create_graph(axes, "sin(x*x)")
```

Функция cos(sin(x)). Аналитически её производная -cos(x)*sin(sin(x))

```python
def function2_derivative(x):
    return (-math.cos(x)*math.sin(math.sin(x)))

axes = [[] for _ in range(6)] #матрица результатов для графика

for n in range(1, 21):
    h = 2 / (2**n)
    axes[0].append(h)
    resultInPoint = function2_derivative(x_0)
    for i in range(0, 5):
        axes[i + 1].append(math.log(abs(f[i](function2, x_0, h) - resultInPoint)))

create_graph(axes, "cos(sin(x))")
```

Функция exp(sin(cos(x))). Аналитически её производная -sin(x) * cos(cos(x)) * exp(sin(cos(x)))

```python
def function3_derivative(x):
    return (-math.sin(x) * math.cos(math.cos(x)) * math.exp(math.sin(math.cos(x))))

axes = [[] for _ in range(6)] #матрица результатов для графика

for n in range(1, 21):
    h = 2 / (2**n)
    axes[0].append(h)
    resultInPoint = function3_derivative(x_0)
    for i in range(0, 5):
        axes[i + 1].append(math.log(abs(f[i](function3, x_0, h) - resultInPoint)))

create_graph(axes, "exp(sin(cos(x)))")
```

Функция ln(x+3). Аналитически её производная (x+3)^-1

```python
def function4_derivative(x):
    return (x+3)**(-1)

axes = [[] for _ in range(6)] #матрица результатов для графика

for n in range(1, 21):
    h = 2 / (2**n)
    axes[0].append(h)
    resultInPoint = function4_derivative(x_0)
    for i in range(0, 5):
        axes[i + 1].append(math.log(abs(f[i](function4, x_0, h) - resultInPoint)))

create_graph(axes, "ln(x+3)")
```

Функция (x+3)**0,5. Аналитически её производная 0,5 * (x+3)^-0,5

```python
def function5_derivative(x):
    return (0.5 * (x + 3)**(-0.5))

axes = [[] for _ in range(6)] #матрица результатов для графика

for n in range(1, 21):
    h = 2 / (2**n)
    axes[0].append(h)
    resultInPoint = function5_derivative(x_0)
    for i in range(0, 5):
        axes[i + 1].append(math.log(abs(f[i](function5, x_0, h) - resultInPoint)))

create_graph(axes, "(x+3)^0.5")
```
